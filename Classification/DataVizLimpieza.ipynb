```python
df=train.drop('CODIGO_EMPRESA',axis=1)
df.describe
```

```python
# Select columns excluding 'FRACASO'
features = df.columns[df.columns != 'FRACASO']

# Set up the plot grid
num_features = len(features)
num_cols = 2  # Assuming you want 2 columns in each row
num_rows = (num_features + num_cols - 1) // num_cols

# Set up the figure
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
fig.suptitle('Box Plots for Features with respect to FRACASO', y=1.02)

# Flatten the axes array for easier iteration
axes = axes.flatten()

for i, feature in enumerate(features):
    # Create a box plot for each feature
    sns.boxplot(x='FRACASO', y=feature, data=df, ax=axes[i])
    axes[i].set_title(f'Box Plot for {feature}')

# Remove any empty subplots if the number of features is not a multiple of num_cols
for j in range(num_features, num_rows * num_cols):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
