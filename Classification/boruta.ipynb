```python
import pandas as pd
import matplotlib.pyplot as plt
from boruta import BorutaPy
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('TrainCsv.csv')
df.head()
```

```python
X=df.iloc[:,2:]
X.head()
y=df.iloc[:,1]
y
```

```python
rf=RandomForestClassifier(class_weight='balanced')
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
feat_selector=BorutaPy(rf, n_estimators='auto',random_state=42)
feat_selector.fit(X.values,y)
```

```python
# zip feature names, ranks, and decisions 
feature_ranks = list(zip(X.columns, 
                         feat_selector.ranking_, 
                         feat_selector.support_))

# print the results
for feat in feature_ranks:
    print('Feature: {:<30} Rank: {},  Keep: {}'.format(feat[0], feat[1], feat[2]))
```

```python
# call transform() on X to filter it down to selected features
selected_features = X.columns[feat_selector.support_].tolist()
selected_features
```

```python
XFiltrado=X.loc[:,feat_selector.support_]
XFiltrado.head()
```
