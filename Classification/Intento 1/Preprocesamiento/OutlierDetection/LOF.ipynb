```python
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
import pandas as pd
```

```python
train=pd.read_csv('C:/Users/Santiago Zubieta/Documents/2023-2/statistical/Proyecto/ProyectoSLDA/Classification/Preprocesamiento/TrainCsv.csv')
train.head()
```

```python
varBasuras=['B11','B13','B14','B21','B23','B24','B46','B48',
            'B51','B52','B53','B54','B61','B62','B63','B64',
            'B65','B66','B67','B611','B85']
X=train.drop(columns=varBasuras)

X.shape
```

```python

X=train.iloc[:,2:]
lof=LocalOutlierFactor(n_neighbors=5,contamination=0.05)
y_pred=lof.fit_predict(X)
```

```python
X.head()
```

```python
outliers=train[y_pred==-1]
outliers
```

```python
remover=outliers[outliers['FRACASO']==0]
remover
```

```python
filtrado=train[~train['CODIGO_EMPRESA'].isin(outliers['CODIGO_EMPRESA'])]
filtrado.shape
```

```python
filtrado['FRACASO'].value_counts()
```

```python
filtrado.to_csv('sinOutliers.csv')
```
