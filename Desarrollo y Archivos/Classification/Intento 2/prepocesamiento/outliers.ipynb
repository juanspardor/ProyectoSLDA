```python
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor
```

```python
train1=pd.read_csv('C:/Users/Santiago Zubieta/Documents/2023-2/statistical/Proyecto/ProyectoSLDA/Classification/intento 1/Preprocesamiento/TrainCsv.csv')
test1=pd.read_csv('C:/Users/Santiago Zubieta/Documents/2023-2/statistical/Proyecto/ProyectoSLDA/Classification/intento 1/Preprocesamiento/TestCsv.csv')


variables2=['B47','B45','B46','B22','B24','B21','B31','B23','B67',
             'B83','B42','B32','B48','B63','B43','B610','B612',
             'B44','B52','B13','B68','B61','B85','B82','B611']
variables2.append('CODIGO_EMPRESA')
test=test1[variables2]
variables2.append('FRACASO')
train=train1[variables2]


tr=train1[variables2]

train.head()
```

```python
X=train
lof=LocalOutlierFactor(n_neighbors=4,contamination=0.05)
y_pred=lof.fit_predict(X)
```

```python
y_pred
```

```python
outliers=train[y_pred==-1]
outliers
```

```python
outliers['FRACASO'].value_counts()
```

```python
remover=outliers[outliers['FRACASO']==0]
remover
```

```python
# Creating a DataFrame with only the selected features


# Fitting the LocalOutlierFactor model
lof = LocalOutlierFactor(contamination=0.05)
outlier_scores = lof.fit_predict(train)

# Creating a boolean mask for outliers with FRACASO equal to 1
mask = (outlier_scores == -1) & (train['FRACASO'] == 0)

# Filtering the DataFrame to remove outliers with FRACASO equal to 1
df_filtered = tr[~mask]
```
